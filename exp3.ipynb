```python
#11
label=[]
tet=[]
for i in range(11):
  strin="/content/drive/MyDrive/txt/"+str(i+1)+".txt"
  f = open(strin)
  text = f.read()
  f.close()
  #print(text)
  label.append(text[:text.find('\n')])
  tet.append(text[text.find('Abstract')+8:text.find('Reference')].replace('\n',' '))
```

```python
dict={'text':tet, 'label':label}
df=pd.DataFrame(dict)
```

```python
df
```

```python
def remove_punc(string):
    punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    for ele in string:  
        if ele in punc:  
            string = string.replace(ele, "") 
    return string
 
df['text']=df['text'].apply(lambda x:remove_punc(x))
```

```python
df['text'] = df['text'].replace(r'[^A-Za-z0-9 ]+', '')

df['text'] = df['text'].apply (lambda x: re.sub(r"((?<=^)|(?<= )).((?=$)|(?= ))", '', x).strip())
```

```python
df['text'][0]
```

```python
stopwords = set(nltk.corpus.stopwords.words('english'))
def lem(line):  #tokenize, stemmed, lemmatize, removed stopwords
  a=[]
  tokenizer = RegexpTokenizer(r'\w+')
  word=tokenizer.tokenize(line)
  lemmatizer = WordNetLemmatizer()

  for each in word:
    lem=lemmatizer.lemmatize(each)
    a.append(lem)

  stemmer_ss = SnowballStemmer("english")  
  words = [stemmer_ss.stem(w) for w in a]  
  final_tokens = []
  for each in words:
    if each not in stopwords:
      final_tokens.append(each)
  return ' '.join(final_tokens)
```

```python
nltk.download('wordnet')
```

```python
df['text']=df['text'].apply(lambda x : lem(x))
#lem(df['text'][0])
```

```python
from nltk.probability import FreqDist
def mf(text):
  tokenizer = RegexpTokenizer(r'\w+')
  word=tokenizer.tokenize(text)  
  fdist = FreqDist(word)
  p=[]
  for i in fdist.keys():
    if fdist[i]>5:
      p.append(i)
  s=' '.join(p)
  return s
```

```python
df['text']=df['text'].apply(lambda x: mf(x))      # have to loop for all 
```
